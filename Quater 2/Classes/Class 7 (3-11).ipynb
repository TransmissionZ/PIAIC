```python
import numpy as np
```

```python
frame = pd.DataFrame({'a': np.random.randn(100)})
store = pd.HDFStore('mydata.h5')
store['obj1'] = frame
store['obj1_col'] = frame['a']
print(store)
x = store.obj1
print(type(x))
print(x.head())
```

```python
frame1 = pd.DataFrame({'a': np.random.randn(100)})
store = pd.HDFStore('mydata1.h5')

store.put('obj2', frame1, format = 'table')
store.select('obj2', where = ['index >= 10 and index <=15'])
store.close()
frame1.to_hdf('mydata1.h5', 'obj3', format= 'table')
frame1 = pd.read_hdf('mydata1.h5', 'obj3', where = ['index<5'])
```
